<h2> Section 2: Initial Hypothesis (or hypotheses) </h2>

<h3>When predicting whether a credit card application will be approved by a particular bank using machine learning classification algorithms, there are several hypotheses and assumptions that need to be made. These assumptions are:
</h3>

<h3>
1. **Representative Data**: The dataset used for training the model is representative of the real-world credit card application data that the bank receives. It includes a diverse range of applicants with various features like income, credit history, age, employment status, etc.
</h3>

<h3>
2. **Feature Relevance**: It is assumed that the features (attributes) used to train the model have some predictive power in determining credit card approval. For instance, features like annual income, occupation type  and employment status are typically considered relevant.
</h3>

<h3>
3. **Data Quality**: It is assumed that the dataset is of high quality, free from errors, and missing values have been handled appropriately (e.g., through imputation or removal of rows with missing data).
</h3>

<h3>
4. **Independence of Observations**: The observations in the dataset are assumed to be independent of each other. In other words, one applicant's approval or rejection should not influence another applicant's decision.
</h3>

<h3>
5. **Stationarity**: The distribution of the data and the relationship between features and approval status are assumed to be relatively stable over time, at least for the duration of the prediction period.
</h3>

<h3>
6. **Class Balance**: It is assumed that the dataset has a reasonably balanced representation of approved and rejected credit card applications. If the classes are heavily imbalanced, it may require special handling techniques like SMOTE(Synthetic Minority OverSampling Technique).
</h3>

<h3>
7. **No Data Leakage**: It is crucial to avoid data leakage, which means unintentionally including information in the training set that would not be available at the time of prediction. For example, using the credit card approval status as a feature would lead to data leakage.
</h3>

<h3>
8. **Model Generalization**: The assumption is that the trained model will generalize well to new, unseen data, and will not overfit to the training set.
</h3>

<h3>
9. **Model Interpretability**: Depending on the regulatory and business requirements, the model's interpretability might be an essential assumption. Some algorithms are inherently more interpretable than others, and this needs to be considered.
</h3>

<h3>
10. **Stability of Decision Criteria**: It is assumed that the criteria used for credit card approval by the bank will remain relatively stable during the deployment of the model. Significant changes in approval policies could affect the model's performance.
</h3>

<h3>
11. **Appropriate Algorithm Choice**: The choice of machine learning classification algorithm is an assumption. Different algorithms have different strengths and weaknesses, and it is essential to select the most appropriate one for the specific task.
</h3>

<h3>Annual income is a good predictor of target variable, checking the quartiles of 'Annual_income' feature
to remove outliers using capping method</h3> 

```python
Q1 = np.percentile(X['Annual_income'], 25, interpolation = 'midpoint')
Q2 = np.percentile(X['Annual_income'], 50, interpolation = 'midpoint')  
Q3 = np.percentile(X['Annual_income'], 75, interpolation = 'midpoint')
```

```python
IQR = Q3 - Q1
print('Interquartile range is', IQR)
low_lim = Q1 - 1.5 * IQR
up_lim = Q3 + 1.5 * IQR
print('low_limit is', low_lim)
print('up_limit is', up_lim)
outlier =[]
for x in X['Annual_income']:
    if ((x> up_lim) or (x<low_lim)):
         outlier.append(x)
print(' outlier in the dataset is', outlier)
```

```python
X['Annual_income'] = X['Annual_income'].clip(lower = low_lim, upper = up_lim)
```

### To find out how much is the correlation among variables/features, I used Variance inflation factor 

```python
# Examine multicollinearity using VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
   
  
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = x_smote.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(x_smote.values, i)
                          for i in range(len(x_smote.columns))]
  
print(vif_data)
```

### Removing/dropping those features which have high variance inflation factor value 

```python
x_smote = x_smote.drop(columns = ['Type_Occupation_Pensioner','Family_Members','Type_Occupation_Working',
                                 'Mobile_phone','Type_Occupation_Working', 'Type_Occupation_Cooking staff','Type_Income_Pensioner'
                                 ,'Type_Income_Working','Type_Income_State servant'])
x_smote.head()
```

```python
# Remaining columns after dropping those variables which have high VIF values
x_smote.columns
```

### Since checking variance inflation factor is an iterative process, so the values of VIF for the features will change after removing/dropping features with high VIF value

```python
vif_data = pd.DataFrame()
vif_data["feature1"] = x_smote.columns
vif_data["VIF1"] = [variance_inflation_factor(x_smote.values, i)
                          for i in range(len(x_smote.columns))]
  
print(vif_data)
```

```python
# Creating a dictionary of models where key is the model name in string format and value is the imported models 
# from their respective libraries 
models={                                   
    "Logistic Regression" :LogisticRegression(max_iter=25000),
    "Decision Tree" :DecisionTreeClassifier(max_features = 'auto', random_state = 44),
    "Random Forest":RandomForestClassifier(n_estimators=10000, random_state = 44),
    "Naive Bayes": GaussianNB(),
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=3),
    "XGBoost Classifier": XGBClassifier(n_estimators = 100, random_state = 123),
    "Catboost Classifier": CatBoostClassifier(iterations = 1500),
    "Gradient Boosting Classifier": GradientBoostingClassifier(),
    "Stochastic Gradient Descent Classifier": SGDClassifier() 
}

for i in range(len(list(models))):   # Iterating through each model in the dictionary and saving into 'model' variable  
    model = list(models.values())[i]
    model.fit(x_smote,y_smote.values.ravel()) # Train Model

    # Make predictions
    y_train_pred = model.predict(x_smote)
    y_test_pred =  model.predict(X_test)

  # Performance of Test set
    model_test_accuracy = accuracy_score(y_test, y_test_pred) 
    model_test_f1 = f1_score(y_test, y_test_pred, average='weighted', zero_division=1) 
    model_test_precision = precision_score(y_test, y_test_pred , average='weighted', zero_division=1) 
    model_test_recall  = recall_score(y_test, y_test_pred,average='weighted', zero_division=1) 
    confusion_matrix_test = metrics.confusion_matrix(y_test, y_test_pred)
    Area_under_curve_test = roc_auc_score(y_test, y_test_pred) 
    classification_report_test = classification_report(y_test, y_test_pred) 

  # Performance of Training set
    model_train_accuracy = accuracy_score(y_smote, y_train_pred) 
    model_train_f1 = f1_score(y_smote, y_train_pred, average= 'weighted', zero_division = 1) 
    model_train_precision = precision_score(y_smote, y_train_pred,average='weighted', zero_division = 1) 
    model_train_recall = recall_score(y_smote, y_train_pred,average='weighted', zero_division = 1) 
    confusion_matrix_train = metrics.confusion_matrix(y_smote, y_train_pred)
    Area_under_curve_train = roc_auc_score(y_smote, y_train_pred)

    print(list(models.keys())[i])

    print('Model performance for Training set')
    print("- Accuracy: {:.2f}".format(model_train_accuracy))
    print('- F1 score: {:2f}'.format(model_train_f1))
    print('- Precision: {:2f}'.format(model_train_precision))
    print('- Recall: {:2f}'.format(model_train_recall))
    print(confusion_matrix_train)
    print('- Area_under_curve_train: {:4f}'.format(Area_under_curve_train))

    print('----------------------------------')

    print('Model performance for Test set')
    print('- Accuracy: {:.2f}'.format(model_test_accuracy) )
    print('- Fl score: {:.2f}'.format(model_test_f1))
    print('- Precision: {:.2f}'.format(model_test_precision))
    print('- Recall: {:.2f}'.format(model_test_recall))
    print(confusion_matrix_test)
    print('- Area_under_curve_test: {:4f}'.format(Area_under_curve_test))
    print(classification_report_test)


    print('='*30)
    print('\n')
```

<h3>Cross validating the models showed that the accuracy value was in fact not arbitary and proofed that the xgboost model is performing well.</h3>

<h3>XGBoost showed a higher mean and lower standard deviation.</h3>

<h3>A high mean corresponds to a more stable performance and a low standard deviation corresponds to smaller range of results.</h3>

<h3> The accuracy score of 84% or 0.84 means that of all the observations/data points the model can classify 84% data points correctly. Out of all the customers in the datset, our model can classify which customer should be chosen for Credit Card and which should not be with 84% accuracy. Since the number of false positives and false negatives are different, we can not rely on accuracy as the only metric.</h3>

<h3>The Precision score of 86% means that out of the total predicted positive observations (i.e, True Positive + False Positive), the model can predict 86% positive observations correctly. 
    
<h3>The Recall score (or sensitivity) of 84% means that out of all the observations in actual class the model can predict the 84% positive observations correctly.

<h3> In this problem of predicting whether the customer should be given credit card or not, precision and recall both should be taken care of alongwith accuracy.</h3>
    
<h3>Since we require both precision and recall, F1 score, which is the harmonic mean of precision and recall should be considered. F1 score of 0.85 which is close to 1 ensures that the model is performing well on test dataset. A higher F1 score signifies that the bank will be able to classify most customers correctly based on their credit worthiness, which customer should be given credit card and which should not be.
